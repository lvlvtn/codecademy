# Visualizing Loans Awarded by Kiva

The Codeacademy Off Platform Project on visualization with Seaborn. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a> downloaded from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. 

Tasks:
- explore the average loan amount by country using aggregated bar charts
- visualize the distribution of loan amounts by project type and gender using box plots and violin plots

## Import Python libraries and download the data

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('kiva_data.csv')
df.head()
```

## Examine The Data

```python
df.head(10)
```

## Simple Bar Charts

```python
import matplotlib.ticker as mtick

# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))

# Plot the data
sns.barplot(data=df, x="country", y = "loan_amount")

# Format the y-axis ticks to add $ (units of USD)
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set(xlabel='Country', ylabel='Loan amount')
```

## Adding `hue` to the Visualization

```python
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount", hue='gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set(xlabel='Country', ylabel='Loan amount')
```

## Styling

```python
# Set color palette
sns.set_palette('Pastel1')

# Set style
sns.set_style('darkgrid')

# Create figure and axes
plt.figure(figsize=(25, 15))

# Add a title
ax.set_title("Loan amount by country and gender", fontsize=12)

# Use Seaborn to create the bar plot
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount", hue='gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

## Box Plots 

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df, x="country", y="loan_amount", hue='gender')
```

## Box Plot by Activity

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df, x="activity", y="loan_amount", hue='gender')
```

## Violin Plots

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y="loan_amount")
```

### Violin plot that visualizes the distribution of loan amount by country

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="country", y="loan_amount")
```

## Split Violin Plots

```python
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x="country", y="loan_amount", hue='gender')
```
